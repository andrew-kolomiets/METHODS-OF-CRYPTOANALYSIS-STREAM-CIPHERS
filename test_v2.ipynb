```python
GB = Ideal(f).groebner_basis()
print(GB)
```

```python
GB_xor_1 =Ideal(f + 1).groebner_basis()
print(GB_xor_1)
```

```python
GB_merging=[*GB,*GB_xor_1]
degrees=[poly.deg() for poly in GB_merging]
AI = min(degrees)

print('AI(f) = ', AI)
```

```python
h1, h2=0, 0
min_deg_poly = []

for poly in GB_merging:
    if poly.deg() == AI:
        min_deg_poly.append(poly)
    
    if(len(min_deg_poly)==2):
        print('h_1 = ', min_deg_poly[0],'\n','h_2 = ', min_deg_poly[1])
        h1=min_deg_poly[0]
        h2=min_deg_poly[1]
```

```python
system = []
for i in range(1000):
    current_state = C * current_state
    if gamma[i] == '0':
        system.append(h1(*(current_state)))
    else:
        system.append(h2(*(current_state)))


system_first_ten_equation=[str(system[i])+"="+str(gamma[i]) for i in range(10)]
print(system_first_ten_equation)
```

```python
GB_system = Ideal(system).groebner_basis()

initial_state =[]

for b in GB_system:
    if(b.has_constant_part()==true):
        initial_state.append(1)
    else:
        initial_state.append(0)
        
        
for i in range(64):
    print(GB_system[i])
```

```python
print('Початковий стан: \n \t',initial_state)
```

```python
current_state = vector(B, initial_state)

found_gamma = ''
while len(found_gamma) < 1000:
    current_state = C*current_state
    temp=f(*current_state)
    found_gamma += str(temp)
```

```python
print( true if found_gamma==gamma[:1000] else false)
```
